```python
import requests
import json
from urllib.parse import urljoin
from pprint import pprint

API = '54262b6457447cafa4023115f5562679'
SECRET = '7f16190e426a3b83cd6ac0850c12c0d9'

ROOT = 'http://ws.audioscrobbler.com/2.0/'
API_URL = f'&api_key={API}&format=json'

def make_url(url: str):
    return urljoin(ROOT, url + API_URL)

CHARTS = make_url('?method=chart.gettoptracks')

def search(track: str, artist: str = None):
    return make_url(f"/2.0/?method=track.search&track={track}&artist={artist}")
```

```python
# MAIN
charts = requests.get(CHARTS)
print(charts)
content = json.loads(charts.content)
```
