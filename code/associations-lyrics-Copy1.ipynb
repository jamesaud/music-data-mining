# Clean Lyrics

```python
import re
import string
from nltk.tokenize import WhitespaceTokenizer
from typing import List

lyrics = res['result']['track']['text']

def clean(word):
    chars = [char for char in word if char not in string.punctuation]
    word = "".join(chars)
    word = word.lower()
    return word    

def tokenize_lyrics(lyrics: str):
    """ Takes lyrics in string format, returns list of words """
    replacement_patterns = (r'\[.*?\]',    # Between Brackets
                            r'\(.*?\)')    # Between Parenthesis

    lyrics_clean = re.sub('|'.join(replacement_patterns), '', lyrics)
    tokenizer = WhitespaceTokenizer()
    tokens = tokenizer.tokenize(lyrics_clean)
    
    for word in tokens:
        if '-' in word:
            tokens += word.split('-')
            tokens.remove(word)
        
    return tokens

tokens = tokenize_lyrics(lyrics)
tokens = map(clean, tokens)
tokens = list(tokens)
```
