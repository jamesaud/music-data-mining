```python
import requests
from pymongo import MongoClient
from pprint import pprint
from typing import List

```

```python
DB = 'all-songs'

# DB CONFIG
client = MongoClient('mongo', 27017)
db = client.music_db
songs_collection = db[DB]
```

```python
SONGS = list(songs_collection.find())
len(SONGS)
```

```python
import json
from IPython.display import clear_output
import random

songs = [song for song in SONGS if song.get('spotify', {}).get('id') and song.get('lyrics')]

def retrieve_gender(name):
    response = requests.get("https://api.genderize.io/?name=" + name)
    return json.loads(response.content)['gender']

def update_db(document_id, gender: str):
    document = songs_collection.find_one_and_update({"_id": document_id}, 
                                         {"$set": {"gender": gender}})
    return document


def insert_genders_in_db(songs):
    i = 0
    for song in songs:
        first_name = song['artist'].split(" ")[0]
        try:
            gender = retrieve_gender(first_name)
        except Exception as e:
            print("Couldn't find gender", e)
            continue
        else:
            song['gender'] = gender
            update_db(song['_id'], gender)
            print(f"Inserted Song {i} " + song['title'])
            i += 1
            clear_output(wait=True)

random.shuffle(songs)
insert_genders_in_db(songs)
```
